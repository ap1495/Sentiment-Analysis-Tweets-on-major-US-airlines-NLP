### NLP and Machine Learning Libraries

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, words
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string, re
```

### Data preprocessing

```python
def normalize_tweets(tweet):
    
    #Replace anything other than letters.
    nopunc_string = re.sub("[^a-zA-Z]", " ", tweet)
    #Tokenizing
    tokens = word_tokenize(nopunc_string)[:]
    result = [word for word in tokens if word.lower() not in stopwords.words('english')]

    #First word is always airline name and we will not be using it.
    #result = result[1:]

    #wnl = WordNetLemmatizer()
    ps = PorterStemmer()
    final = []
    for word in result:
        #final.append(wnl.lemmatize(word))
        final.append(ps.stem(word))

    
    return ' '.join(final)

def sentiment_converter(senti):
    if senti == 'negative':
        return 0
    
    elif senti == 'neutral':
        return 1
    
    else:
        return 2
```

```python
df['NormalizedTweet'] = df['text'].apply(normalize_tweets)
df['airline_sentiment'] = df['airline_sentiment'].apply(sentiment_converter)
```

```python
df[['text', 'NormalizedTweet', 'airline_sentiment']].sample(5)
```

### Count Vectorizer

```python
X = df['NormalizedTweet']
y = df['airline_sentiment']
```

```python
cv = CountVectorizer(ngram_range=(1,3))
X_cv = cv.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_cv, y, test_size=0.33, random_state=42)
```

### SGDClassifer

```python
sgdc = SGDClassifier()
sgdc.fit(X_train, y_train)
```

```python
sgdc1_predictions = sgdc.predict(X_test)
print(classification_report(y_test, sgdc1_predictions), '\n')
print(confusion_matrix(y_test, sgdc1_predictions), '\n')
print(accuracy_score(y_test, sgdc1_predictions), '\n')
```

The model is able to better classify **negative** sentiment. It does reasonably well on **neutral** and **positive** sentiments. 

### Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr1_pred = lr.predict(X_test)

print(classification_report(y_test, lr1_pred), '\n')
print(confusion_matrix(y_test, lr1_pred), '\n')
print(accuracy_score(y_test, lr1_pred), '\n')
```

Logistic Regression model does an overall better job than SGDClassifier but the classification of **neutral** tweets goes down while classification of **positive** tweets increases.

### TFIDF Transformer

```python
tf = TfidfTransformer()

X_tfidf = tf.fit_transform(X_cv)

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.33, random_state=42)
```

### SGDClassifier

```python
sgdc.fit(X_train, y_train)
```

```python
sgdc2_predictions = sgdc.predict(X_test)
print(classification_report(y_test, sgdc2_predictions), '\n')
print(confusion_matrix(y_test, sgdc2_predictions), '\n')
print(accuracy_score(y_test, sgdc2_predictions), '\n')
```

After applying the tfidf transformer, the overall performance of the SGDClassifier goes down but the classification of **negative** tweets get better.

### Logistic Regression

```python
lr.fit(X_train, y_train)
```

```python
lr2_pred = lr.predict(X_test)

print(classification_report(y_test, lr2_pred), '\n')
print(confusion_matrix(y_test, lr2_pred), '\n')
print(accuracy_score(y_test, lr2_pred), '\n')
```

The same can be said for this logistic regression model. Overall performance drastically reduces but classification of **negative** tweets get better.<br><br>

## Overall, using the CountVectorizer with ngram_range of (1,3) and a Logistic Regression model, we get an overall accuracy of - 79.59%
